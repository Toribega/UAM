```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV  
df = pd.read_csv('/content/ Raw Data-1 (1).csv', encoding='utf-8')

#  
df.rename(columns={'1.1	          ': '1',
                   '1.2	  ,  ,       ': '2',
                   '1.3	        ': '3',
                   '1.4	       ': '4',
                   '1.5	      ?': '5',
                   '3.1	     ': '_1',
                   '3.2	       ': '_2',
                   '1.1	UAM   ': '',
                   '2.1	     UAM  ': '1',
                   '2.2	   UAM  ': '2',
                   '4.3	 UAM   ': '3',
                   '4.4	 UAM   ': '4',
                   '5.1	      ': '',
                   '1. UAM   1km X 3000  UAM   ?': '3000',
                   '2. UAM   1km X 2000  UAM   ?': '2000',
                   '3. UAM   1km X 1300  UAM   ?': '1300',
                   '2.    ?': '',
                   '5.      ?': '',
                   '6.  ()   ??': '',
                   '7.     ?': '',
                   '8.        ?': ''}, inplace=True)

#    
df['_2'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df['1'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['2'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['3'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['4'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['5'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df[''].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df['1'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['2'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df[''].replace({1: '10', 2: '20', 3: '30', 4: '40',
                   5: '50', 6: '60', 7: '70'}, inplace=True)


#   
age_order = ['10', '20', '30', '40', '50', '60', '70']
# 5  
order_list = [' ', ' ', '', ' ', ' ']

plt.figure(figsize=(15, 25))

#        
for i, col in enumerate(['1', '2', '3', '4', '5', '', '1', '2'], start=1):
    plt.subplot(4, 2, i)
    ax = sns.countplot(data=df, x='', hue=col, palette='viridis', order=age_order, hue_order=order_list)

    #   
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

    plt.title(f' {col} ')
    plt.xlabel('')
    plt.ylabel('')
    plt.legend(title=col, loc='upper right')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
