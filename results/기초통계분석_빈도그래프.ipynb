```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = "NanumGothic"

# CSV 파일 불러오기
df = pd.read_csv('/content/설문조사결과 Raw Data-1 (1).csv', encoding='utf-8')

# 변수명 수정
df.rename(columns={'1.1	나는 여행을 통해 일상에서 잠시 휴식을 취하고 싶은 욕구를 자주 느낀다': '관광동기1',
                   '1.2	고급 숙박 시설, 고급 식사, 특별한 경험의 매력이 여행 선택에 중요한 부분이다': '관광동기2',
                   '1.3	여행을 하면서 새롭고 다양한 경험을 찾는 것에 관심이 있다': '관광동기3',
                   '1.4	변화에 대한 열망과 활력이 여행 결정에 영향을 미친다': '관광동기4',
                   '1.5	여행 중에 자연과 연결되는 것이 얼마나 중요합니까?': '관광동기5',
                   '3.1	관광지 간 이동은 여행의 중요한 부분이다': '관광지_이동수단1',
                   '3.2	관광지 간 이동수단의 편리성은 여행 만족도에 영향이 크다': '관광지_이동수단2',
                   '1.1	UAM이 나에게 유용할 것이다': '유용성',
                   '2.1	나는 흥미와 즐거움을 느끼기 위해 UAM을 사용할 것이다': '사용의도1',
                   '2.2	나는 호기심이 많아 UAM을 사용해보고 싶다': '사용의도2',
                   '4.3	나는 UAM이 무엇인지 알고 있다': '사용의도3',
                   '4.4	나는 UAM에 대해서 관심이 많다': '사용의도4',
                   '5.1	나는 혁신적인 새로운 기술에 대해서 관심이 많다': '수용태도',
                   '1. UAM 이용 금액이 1km X 3000원이라고 할때 UAM을 탑승할 의향이 있으십니까?': '이용금액3000',
                   '2. UAM 이용 금액이 1km X 2000원이라고 할때 UAM을 탑승할 의향이 있으십니까?': '이용금액2000',
                   '3. UAM 이용 금액이 1km X 1300원이라고 할때 UAM을 탑승할 의향이 있으십니까?': '이용금액1300',
                   '2. 귀하의 연령은 어떻게 되십니까?': '연령',
                   '5. 귀하의 개인 소득 수준은 어떻게 되십니까?': '소득',
                   '6. 항공기 탑승(년) 횟수는 어떻게 되십니까??': '항공기탑승횟수',
                   '7. 주로 이용하는 교통수단은 어떻게 되십니까?': '주교통수단',
                   '8. 귀하의 거주지역의 교통수단 보급의 체감 정도는 어떻게 되십니까?': '보급체감정도',
                   '7.1	운송 옵션을 선택할 때 환경 지속 가능성은 중요하다': '환경친화성'}, inplace=True)

# 변수 값에 명칭 부여
df['관광지_이동수단2'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)

df['관광동기1'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)
df['관광동기2'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)
df['관광동기3'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)
df['관광동기4'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)
df['관광동기5'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)

df['유용성'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)

df['사용의도1'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)
df['사용의도2'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)

df['환경친화성'].replace({1: '전혀 아니다', 2: '약간 아니다', 3: '보통이다', 4: '약간 그렇다',
                   5: '매우 그렇다'}, inplace=True)

df['이용금액3000'].replace({1: '예', 2: '아니오', 3: '잘 모르겠다'}, inplace=True)
df['이용금액2000'].replace({1: '예', 2: '아니오', 3: '잘 모르겠다'}, inplace=True)
df['이용금액1300'].replace({1: '예', 2: '아니오', 3: '잘 모르겠다'}, inplace=True)


df['연령'].replace({1: '10대', 2: '20대', 3: '30대', 4: '40대',
                   5: '50대', 6: '60대', 7: '70대이상'}, inplace=True)


# 연령 순서 정의
age_order = ['10대', '20대', '30대', '40대', '50대', '60대', '70대이상']
# 5점척도 순서 정의
order_list = ['전혀 아니다', '약간 아니다', '보통이다', '약간 그렇다', '매우 그렇다']
# 이용금액 변수 값에 대한 순서 정의
price_order = ['예', '아니오', '잘 모르겠다']

# 첫 번째 그래프: 연령별 관광동기 및 사용의도에 대한 빈도
plt.figure(figsize=(15, 60))
for i, col in enumerate(['관광동기1', '관광동기2', '관광동기3', '관광동기4', '관광동기5', '유용성', '사용의도1', '사용의도2', '환경친화성'], start=1):
    plt.subplot(9, 1, i)
    ax = sns.countplot(data=df, x='연령', hue=col, palette='viridis', order=age_order, hue_order=order_list)
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=16, color='black', xytext=(0, 5), textcoords='offset points')
    plt.title(f'연령별 {col} 빈도', fontsize=30)
    plt.xlabel('연령', fontsize=25)
    plt.ylabel('빈도', fontsize=25)
    plt.legend(title=col, loc='upper right', bbox_to_anchor=(1.0, 1), title_fontsize='xx-large', fontsize='xx-large')
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()


# 두 번째 그래프: 연령별 이용금액에 대한 빈도
plt.figure(figsize=(15, 20))
for i, col in enumerate(['이용금액3000', '이용금액2000', '이용금액1300'], start=1):
    plt.subplot(3, 1, i)
    ax = sns.countplot(data=df, x='연령', hue=col, palette='viridis', order=age_order, hue_order=price_order)
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=18, color='black', xytext=(0, 5), textcoords='offset points')
    plt.title(f'연령별 {col} 빈도', fontsize=30)
    plt.xlabel('연령', fontsize=25)
    plt.ylabel('빈도', fontsize=25)
    plt.legend(title=col, loc='upper right', bbox_to_anchor=(1.0, 1), title_fontsize='xx-large', fontsize='xx-large')
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# 필요한 열을 선택하여 새로운 DataFrame을 만듭니다.
selected_columns = ['연령', '관광동기1', '관광동기2', '관광동기3', '관광동기4', '관광동기5', '유용성', '사용의도1', '사용의도2', '환경친화성', '이용금액3000', '이용금액2000', '이용금액1300']
df_selected = df[selected_columns]

# '연령' 열을 기준으로 그룹화하여 다른 열의 통계를 계산하고 출력합니다.
for col in selected_columns[1:]:
    grouped_counts = df_selected.groupby('연령')[col].value_counts().unstack(fill_value=0)
    print(f"=== {col} ===")
    display(grouped_counts)
    print()
```
