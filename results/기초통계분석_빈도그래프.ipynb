```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = "NanumGothic"

# CSV  
df = pd.read_csv('/content/ Raw Data-1 (1).csv', encoding='utf-8')

#  
df.rename(columns={'1.1	          ': '1',
                   '1.2	  ,  ,       ': '2',
                   '1.3	        ': '3',
                   '1.4	       ': '4',
                   '1.5	      ?': '5',
                   '3.1	     ': '_1',
                   '3.2	       ': '_2',
                   '1.1	UAM   ': '',
                   '2.1	     UAM  ': '1',
                   '2.2	   UAM  ': '2',
                   '4.3	 UAM   ': '3',
                   '4.4	 UAM   ': '4',
                   '5.1	      ': '',
                   '1. UAM   1km X 3000  UAM   ?': '3000',
                   '2. UAM   1km X 2000  UAM   ?': '2000',
                   '3. UAM   1km X 1300  UAM   ?': '1300',
                   '2.    ?': '',
                   '5.      ?': '',
                   '6.  ()   ??': '',
                   '7.     ?': '',
                   '8.        ?': '',
                   '7.1	       ': ''}, inplace=True)

#    
df['_2'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df['1'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['2'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['3'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['4'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['5'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df[''].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df['1'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)
df['2'].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df[''].replace({1: ' ', 2: ' ', 3: '', 4: ' ',
                   5: ' '}, inplace=True)

df['3000'].replace({1: '', 2: '', 3: ' '}, inplace=True)
df['2000'].replace({1: '', 2: '', 3: ' '}, inplace=True)
df['1300'].replace({1: '', 2: '', 3: ' '}, inplace=True)


df[''].replace({1: '10', 2: '20', 3: '30', 4: '40',
                   5: '50', 6: '60', 7: '70'}, inplace=True)


#   
age_order = ['10', '20', '30', '40', '50', '60', '70']
# 5  
order_list = [' ', ' ', '', ' ', ' ']
#      
price_order = ['', '', ' ']

#   :      
plt.figure(figsize=(15, 60))
for i, col in enumerate(['1', '2', '3', '4', '5', '', '1', '2', ''], start=1):
    plt.subplot(9, 1, i)
    ax = sns.countplot(data=df, x='', hue=col, palette='viridis', order=age_order, hue_order=order_list)
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=16, color='black', xytext=(0, 5), textcoords='offset points')
    plt.title(f' {col} ', fontsize=30)
    plt.xlabel('', fontsize=25)
    plt.ylabel('', fontsize=25)
    plt.legend(title=col, loc='upper right', bbox_to_anchor=(1.0, 1), title_fontsize='xx-large', fontsize='xx-large')
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()


#   :    
plt.figure(figsize=(15, 20))
for i, col in enumerate(['3000', '2000', '1300'], start=1):
    plt.subplot(3, 1, i)
    ax = sns.countplot(data=df, x='', hue=col, palette='viridis', order=age_order, hue_order=price_order)
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=18, color='black', xytext=(0, 5), textcoords='offset points')
    plt.title(f' {col} ', fontsize=30)
    plt.xlabel('', fontsize=25)
    plt.ylabel('', fontsize=25)
    plt.legend(title=col, loc='upper right', bbox_to_anchor=(1.0, 1), title_fontsize='xx-large', fontsize='xx-large')
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()
```

```python
#     DataFrame .
selected_columns = ['', '1', '2', '3', '4', '5', '', '1', '2', '', '3000', '2000', '1300']
df_selected = df[selected_columns]

# ''        .
for col in selected_columns[1:]:
    grouped_counts = df_selected.groupby('')[col].value_counts().unstack(fill_value=0)
    print(f"=== {col} ===")
    display(grouped_counts)
    print()
```
